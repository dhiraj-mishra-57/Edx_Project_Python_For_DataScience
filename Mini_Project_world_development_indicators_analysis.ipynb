## In This Notebook i have done analysis on World-Development-Indicators Dataset##

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./world-development-indicators/Indicators.csv')
data.head()
```

### Sample Commands gives you some random data in DataFrame###

```python
data.sample(30)
```

```python
hospital_data = data['IndicatorName'].str.contains('Hospital beds \(per 1,000 people\)')
hospital_data = data[hospital_data]
hospital_data.head()
```

```python
hospital_data.info()
```

```python
majorcountry_code = ['AUS','CAN','CHN','DEU','IND','ISR','JPN','RUS','GBR','USA']
len(majorcountry_code)
```

```python
#df_1 = df['CountryCode'].majorcountry_code
#help(df.isin)
develop_hospital_data = hospital_data['CountryCode'].isin(majorcountry_code)
develop_hospital_data = hospital_data[develop_hospital_data]
print(hospital_data.shape)
print(develop_hospital_data.shape)
```

```python
develop_hospital_data.sample(10)
```

```python
develop_hospital_data = develop_hospital_data.sort_values(by='Year')
develop_hospital_data.head()
```

```python
develop_hospital_data_2010 = (develop_hospital_data['Year'] == 2010)
develop_hospital_data_2010 = develop_hospital_data[develop_hospital_data_2010]
develop_hospital_data_2010.head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
fig, ax = plt.subplots()
fig.set_size_inches(18, 7)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
sns.pointplot(x=develop_hospital_data_2010['CountryName'], y=develop_hospital_data_2010['Value'],data=develop_hospital_data_2010,join=True)
ax.set(xlabel='Country Names', ylabel=develop_hospital_data_2010['IndicatorName'].iloc[1],
       title="Hospital beds based of Developed Countries in 2010")
```

## Below is the Code in which we are analyzing the population Growth of India ##

```python
!ls ./world-development-indicators/
```

```python
df = pd.read_csv('./world-development-indicators/Indicators.csv')
df.head()
```

```python
population_data = df['IndicatorName'].str.contains('Population, total')
population_data = df[population_data]
population_data.head()
```

```python
population_data.info()
```

```python
#Setting up the country code of India
code = ['IND']
```

```python
population_data_india = population_data['CountryCode'].isin(code)
population_data_india = population_data[population_data_india]
population_data_india.head()
len(population_data_india)
```

```python
population_data_india.info()
```

```python
population_data_india.head()
```

```python
population_data_india['new_value'] = (population_data_india['Value'].values /100000000)
```

```python
population_data_india.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
fig, ax = plt.subplots()
fig.set_size_inches(18, 7)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
sns.pointplot(x=population_data_india['Year'], y=population_data_india['new_value'],data=population_data_india,join=True)
ax.set(xlabel='Years from 1960 to 2014', ylabel='Population in values of e^8',
       title='India\'s Population over the Years')
```
